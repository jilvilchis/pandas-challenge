
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#setting float format more cosmetic
pd.options.display.float_format = '${:,.2f}'.format

#Running calculations and assigning them to a variables
male = purchase_data.loc[(purchase_data["Gender"]=="Male")]
female = purchase_data.loc[(purchase_data["Gender"]=="Female")]
other = purchase_data.loc[(purchase_data["Gender"]=="Other / Non-Disclosed")]

male_purchase = male["Purchase ID"].count()
female_purchase = female["Purchase ID"].count()
other_purchase = other["Purchase ID"].count()

male_avgprice = male["Price"].mean()
female_avgprice = female["Price"].mean()
other_avgprice = other["Price"].mean()

male_sum = male["Price"].sum()
female_sum = female["Price"].sum()
other_sum = other["Price"].sum()

unique_males = male["SN"].nunique()
unique_females = female["SN"].nunique()
unique_other = other["SN"].nunique()
Total = purchase_data["SN"].nunique()

male_avgperson = male["Price"].sum() / unique_males 
female_avgperson = female["Price"].sum() / unique_females
other_avgperson = other["Price"].sum() / unique_other

#Creating and displaying a summary dataframe
summary_purchasing = pd.DataFrame({
    "Gender":["Male","Female","Other/Non-Disclosed"],
    "Purchase Count":[male_purchase,female_purchase,other_purchase],
    "Average Purchase Price": [male_avgprice,female_avgprice,other_avgprice],
    "Total Purchase Value": [male_sum,female_sum,other_sum],
    "Avg Total Purchase Per Person": [male_avgperson,female_avgperson,other_avgperson],
    })

summary_purchasing = summary_purchasing.set_index("Gender")
summary_purchasing
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#setting float format more cosmetic
pd.options.display.float_format = '{:,.2%}'.format

#Creating bins in which to place values and bin names
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
biname = ["<10","10-14","15-19","20-24","25-29","30-34","35-39", "40+"]

#Creating a Dataframe with unique SN to avoid overstate and binning data
Age_demo = purchase_data.copy().drop_duplicates("SN") 
Age_demo["Age Demographics"] = pd.cut(Age_demo["Age"],bins=bins, labels=biname) 
Age_demo

#Creating a dataframe grouping by Age Demographcis and using the count method
Age_bin = Age_demo[["Age Demographics","Age"]].groupby("Age Demographics").count().reset_index()

#Creating the "Percentage of Players" column and adding it to Age:bin dataframe
suma = Age_bin["Age"].sum()
Age_percent = [Age_bin.iloc[0,1]/suma, Age_bin.iloc[1,1]/suma, Age_bin.iloc[2,1]/suma, 
               Age_bin.iloc[3,1]/suma, Age_bin.iloc[4,1]/suma, Age_bin.iloc[5,1]/suma,
               Age_bin.iloc[6,1]/suma, Age_bin.iloc[7,1]/suma]
Age_bin["Percentage of Players"] = Age_percent 

#Renaming columns
Age_bin = Age_bin.rename(columns={"Age":"Total Count"})
Age_bin = Age_bin.set_index("Age Demographics")
Age_bin
```

```python
#To reset the index on Age_bin dataframe as we are going to use it again
Age_bin = Age_bin.reset_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#setting float format more cosmetic
pd.options.display.float_format = '${:,.2f}'.format

#Creating bins in which to place values and bin names
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
biname = ["<10","10-14","15-19","20-24","25-29","30-34","35-39", "40+"]

#Creating a Dataframe with non-unique SN
Age_demo_total = purchase_data.copy()
Age_demo_total ["Age Demographics"] = pd.cut(Age_demo_total["Age"],bins=bins, labels=biname) 
Age_demo_total

#Creating a dataframe grouping by Age Demographcis and using the count method over Purchase ID
Purch_count = Age_demo_total[["Age Demographics","Purchase ID"]].groupby("Age Demographics").count().reset_index()
Purch_count

#Creating a dataframe grouping by Age Demographcis and using the mean method over Price
Purch_Avg = Age_demo_total[["Age Demographics","Price"]].groupby("Age Demographics").mean().reset_index()
Purch_Avg

#Creating a dataframe grouping by Age Demographcis and using the sum method over Price
Purch_Sum = Age_demo_total[["Age Demographics","Price"]].groupby("Age Demographics").sum().reset_index()
Purch_Sum

#Merging 3 prior dataframes
merge_df = pd.merge(Purch_count, Purch_Avg, on="Age Demographics", how="inner")
merge_df = pd.merge(merge_df, Purch_Sum, on="Age Demographics", how="inner")

#Creating the "Avg Total Purchase per Person" column and adding it to consolidated dataframe
Tavg_purch_person = [Purch_Sum.iloc[0,1]/Age_bin.iloc[0,1], Purch_Sum.iloc[1,1]/Age_bin.iloc[1,1],
                     Purch_Sum.iloc[2,1]/Age_bin.iloc[2,1], Purch_Sum.iloc[3,1]/Age_bin.iloc[3,1],
                     Purch_Sum.iloc[4,1]/Age_bin.iloc[4,1], Purch_Sum.iloc[5,1]/Age_bin.iloc[5,1],
                     Purch_Sum.iloc[6,1]/Age_bin.iloc[6,1], Purch_Sum.iloc[7,1]/Age_bin.iloc[7,1]]
merge_df["Avg Total Purchase per Person"] = Tavg_purch_person 

#Renaming columns
merge_df = merge_df.rename(columns={"Age Demographics":"Age Ranges", 
                                    "Purchase ID":"Purchase Count",
                                    "Price_x":"Average Purchase Price", 
                                    "Price_y":"Total Purchase Value"})
merge_df = merge_df.set_index("Age Ranges")
merge_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#setting float format more cosmetic
pd.options.display.float_format = '${:,.2f}'.format

#Creating a dataframe grouping by SN and using the count method over Price
Top5_Count = purchase_data[["SN", "Price"]].groupby("SN").count().reset_index()

#Creating a dataframe grouping by SN and using the mean method over Price
Top5_avg = purchase_data[["SN", "Price"]].groupby("SN").mean().reset_index()

#Creating a dataframe grouping by SN and using the sum method over Price
Top5_sum = purchase_data[["SN", "Price"]].groupby("SN").sum().reset_index()

#Merging 3 prior dataframes
merge_top = pd.merge(Top5_Count, Top5_avg, on="SN", how="inner")
merge_top = pd.merge(merge_top, Top5_sum, on="SN", how="inner")

#Sorting merged df by Total Purchase Price
merge_top = merge_top.sort_values("Price", ascending = False )

#Renaming columns
merge_top = merge_top.rename(columns={"Price_x":"Purchase Count", 
                                    "Price_y":"Average Purchase Price",
                                    "Price":"Total Purchase Value"})
merge_top = merge_top.set_index("SN")
merge_top.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#setting float format more cosmetic
pd.options.display.float_format = '${:,.2f}'.format

#Retrieveing the Item, Item Name and Item Price columns from original data frame
most_pop= purchase_data[["Item ID", "Item Name"]].copy().drop_duplicates("Item ID")

#Grouping by Item ID and using the count method over Item Name
most_pop_count = purchase_data[["Item ID", "Item Name"]].groupby("Item ID").count().reset_index()

#Grouping by Item ID and using the mean method over Item price
most_pop_mean = purchase_data[["Item ID", "Price"]].groupby("Item ID").mean().reset_index()

#Grouping by Item ID and using the sum method over Item price
most_pop_sum = purchase_data[["Item ID", "Price"]].groupby("Item ID").sum().reset_index()

#Merging prior dataframes
merge_most = pd.merge(most_pop, most_pop_count, on="Item ID", how="inner")
merge_most = pd.merge(merge_most, most_pop_mean,  on="Item ID", how="inner")
merge_most = pd.merge(merge_most, most_pop_sum, on="Item ID", how="inner")

#Sorting merged df by Total Purchase Price
merge_most = merge_most.sort_values("Item Name_y", ascending = False )

#Renaming columns
merge_most = merge_most.rename(columns={"Item Name_x":"Item Name", 
                                    "Item Name_y":"Purchase Count",
                                    "Price_x":"Item Price",
                                    "Price_y":"Total Purchase Value"})
merge_most = merge_most.set_index("Item ID")
merge_most.head(10)
```
